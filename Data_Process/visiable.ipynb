```python
import pandas as pd
```

```python
df = pd.read_csv('asian/asian_museum.csv')
df
```

```python
df_filtered = df[
    ['Credit Line', 'Materials', 'Artist',  'Date', 'Classifications',
    'Dynasty', 'More Information', 'Dimensions', 'Title', 'Place of Origin']]
# df_filtered = df[
#     ['Img_path', 'Credit Line', 'Materials', 'Artist', 'Markings', 'On View', 'Date', 'Inscribed', 'Classifications',
#      'Signed', 'Dynasty', 'More Information', 'Img_url', 'Dimensions', 'Title', 'Place of Origin']]
df_filtered.rename(columns={'More Information': 'Description'}, inplace=True)
df_filtered
```

```python
df_filtered.to_csv('datasets_museum/asian_tt.csv', index=False)
```
